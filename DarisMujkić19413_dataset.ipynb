```python
def get_all_links_from_subcategory(subcategory_url):
    all_links = set()
    page_num = 1

    while True:
        page_url = f"{subcategory_url}?page={page_num}"
        print("Scraping:", page_url)
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
        }
        response = requests.get(page_url, headers=headers)
        if response.status_code != 200:
            print("Status kod:", response.status_code)
            break

        soup = BeautifulSoup(response.text, "html.parser")

        # Nađi sve div-ove koji sadrže članke
        article_divs = soup.find_all("div", class_="position-relative mb-30")
        if not article_divs:
            print("Nema više članaka na stranici.")
            break

        broj_linkova_na_stranici = 0

        for div in article_divs:
            h3 = div.find("h3")
            if h3:
                a_tag = h3.find("a", href=True)
                if a_tag:
                    full_link = a_tag["href"]
                    if not full_link.startswith("http"):
                        full_link = "https://startbih.ba" + full_link
                    if full_link not in all_links:
                        all_links.add(full_link)
                        broj_linkova_na_stranici += 1

        print(f"Scraping stranice sa URL-om: {page_url}")
        print(f"Stranica {page_num} — {broj_linkova_na_stranici} članaka")
        print(f"Do sad ukupno linkova: {len(all_links)}")

        if broj_linkova_na_stranici == 0:
            break

        page_num += 1
        time.sleep(random.uniform(0.1, 0.4))

    return list(all_links)
```

```python
def scrape_and_save(category_name, subcategory_links):
    all_links = set()

    for sub_url in subcategory_links:
        links = get_all_links_from_subcategory(sub_url)
        all_links.update(links)

    # Definiši putanju na Google Drive
    filename = f"{category_name}.txt"
    filepath = os.path.join(save_dir, filename)

    # Spremi u txt fajl direktno na Drive
    with open(filepath, "w", encoding="utf-8") as f:
        for link in sorted(all_links):
            f.write(link + "\n")

    print(f"[{category_name}] Ukupno spašeno linkova: {len(all_links)} u fajl: {filepath}")
```

```python
article_url = "https://startbih.ba/clanak/34-godisnjak-poginuo-u-prometnoj-nesreci-kod-jajca/121524"  # zamijeni s pravim linkom

def parse_article(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # -------------------------------
    # Naslov iz <h2> unutar .news-details-layout1
    title_tag = soup.select_one("div.news-details-layout1 h2.title-semibold-dark.size-c30")
    naslov = title_tag.text.strip() if title_tag else "N/A"

    # -------------------------------
    # Podnaslov i tekst iz .editor mb-30
    editor_div = soup.select_one("div.editor.mb-30")
    podnaslov = "N/A"
    tekst_lines = []

    if editor_div:
        all_paragraphs = editor_div.find_all("p", recursive=False)
        first_p = True
        for p in all_paragraphs:
            # Preskoči prazne i &nbsp; paragrafe
            content = p.get_text(strip=True)
            if not content or content == "\xa0":
                continue

            if first_p:
                podnaslov = content
                first_p = False
            else:
                tekst_lines.append(content)

    tekst = "\n".join(tekst_lines) if tekst_lines else "N/A"

    # -------------------------------
    # Rubrika iz URL-a (npr. /vijesti/)
    try:
        rubrika = url.split("/clanak/")[0].rstrip("/").split("/")[-1].upper()
    except:
        rubrika = "N/A"

    # -------------------------------
    # Autor iz <meta> ili JSON-LD
    autor = "N/A"
    datum = "N/A"

    # Meta tag
    author_meta = soup.find("meta", {"property": "article:author"})
    if author_meta and author_meta.get("content"):
        autor = author_meta["content"]

    # JSON-LD <script>
    ld_json = soup.find("script", type="application/ld+json")
    if ld_json:
        try:
            ld_data = json.loads(ld_json.string)
            autor = ld_data.get("author", {}).get("name", autor)
            datum = ld_data.get("datePublished", "N/A")
        except json.JSONDecodeError:
            pass

    # -------------------------------
    # Finalni structured format
    article_data = {
        "NOVINA": "StartBiH.ba",
        "DATUM": datum,
        "RUBRIKA": rubrika,
        "NADNASLOV": "N/A",
        "NASLOV": naslov,
        "PODNASLOV": podnaslov,
        "STRANA": "N/A",
        "AUTOR(I)": autor,
        "TEKST": tekst
    }

    return article_data

# Spremanje u JSON i generisanje .txt
def save_article_to_txt(article_data, output_txt_path, output_json_path):
    if article_data is None:
        print("Nema podataka za spremiti.")
        return

    # Save JSON
    with open(output_json_path, "w", encoding="utf-8") as jf:
        json.dump(article_data, jf, ensure_ascii=False, indent=2)

    # Save TXT u traženom formatu
    with open(output_txt_path, "w", encoding="utf-8") as tf:
        tf.write("<***>\n")
        tf.write(f"* NOVINA: {article_data['NOVINA']}\n")
        tf.write(f"* DATUM: {article_data['DATUM']}\n")
        tf.write(f"* RUBRIKA: {article_data['RUBRIKA']}\n")
        tf.write(f"* NADNASLOV: {article_data['NADNASLOV']}\n")
        tf.write(f"* NASLOV: {article_data['NASLOV']}\n")
        tf.write(f"* PODNASLOV: {article_data['PODNASLOV']}\n")
        tf.write(f"* STRANA: {article_data['STRANA']}\n")
        tf.write(f"* AUTOR(I): {article_data['AUTOR(I)']}\n")
        tf.write("\n")
        tf.write(article_data["TEKST"])
        tf.write("\n\n")

# === Testiranje na jednom članku ===
article_data = parse_article(article_url)

# Putanje za spremanje
json_path = "/content/drive/MyDrive/startbih_parsed/test_article.json"
txt_path = "/content/drive/MyDrive/startbih_parsed/test_article.txt"

save_article_to_txt(article_data, txt_path, json_path)

print("Spremanje završeno.")
```

```python
import os
import shutil
import json
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import re
import time
import html # Za html.unescape
import concurrent.futures # Za ThreadPoolExecutor

base_input_dir = '/content/drive/MyDrive/startbih_links'
base_output_dir = '/content/drive/MyDrive/startbih_parsed'

kategorije = [
    'Pozitivna_BiH', 'Intervju', 'Kolumne', 'Magazin',
    'Politika', 'Ekonomija', 'Sport', 'Vijesti', 'Zivot'
]
NOVINA_NAZIV = "Start BiH"
REQUEST_TIMEOUT = 10 # Sekunde
#PAUZA_IZMEDJU_ZAHTJEVA = 0.5 # Sekunda, da ne opteretimo server
MAX_WORKERS_THREADS = 10

# Prefiksi za ignoriranje paragrafa u sadržaju članka
PREFIXES_TO_IGNORE_IN_CONTENT = (
    'foto:', 'preuzeto:', 'izvor:', 'priredila:', 'priredio:',
    'prevela:', 'preveo:', 'razgovarao:', 'razgovarala:',
    'prevod:', 'prenosi:', 'autor:', 'autori:', 'video:',
    'čitajte još:', 'povezane vijesti:', 'ključne riječi:',
    'tagovi:', 'pratite nas na', 'sva prava pridržana',
    'zabranjeno preuzimanje sadržaja', 'kliknite za više informacija',
    'reklama', 'oglas', 'promo', '(promo)', 'sponzorirani članak',
    'tekst se nastavlja ispod oglasa', 'nastavak teksta ispod oglasa',
    'pratite putem aplikacija za', 'pridružite nam se na viberu'
)
# Lista generičkih imena autora koje "Piše:" treba da zamijeni
GENERIC_AUTHOR_NAMES = ["n/a", "redakcija", "startbih", "startbih.ba", "fena", "srna", "onasa", "agencije", "desk"]

def kreiraj_foldere_i_fajlove():
    if os.path.exists(base_output_dir):
        print(f"Upozorenje: Folder '{base_output_dir}' već postoji. Brišem ga radi čistog početka.")
        try:
            shutil.rmtree(base_output_dir)
        except OSError as e:
            print(f"Greška pri brisanju foldera {base_output_dir}: {e}")
            return False # Prekini ako ne možeš obrisati

    os.makedirs(base_output_dir, exist_ok=True)
    print(f"Kreiran bazni izlazni folder: {base_output_dir}")

    for kategorija_folder_naziv in kategorije:
        folder_path = os.path.join(base_output_dir, kategorija_folder_naziv)
        os.makedirs(folder_path, exist_ok=True)

        # Kreiranje praznih .txt i .json datoteka (bit će prepisane kasnije)
        txt_path = os.path.join(folder_path, f"{kategorija_folder_naziv}.txt")
        json_path = os.path.join(folder_path, f"{kategorija_folder_naziv}.json")

        with open(txt_path, 'w', encoding='utf-8') as f:
            pass
        with open(json_path, 'w', encoding='utf-8') as f:
            json.dump([], f, ensure_ascii=False, indent=4)

    print("Svi folderi i inicijalne datoteke su uspješno kreirani/pripremljeni.")
    return True

kreiraj_foldere_i_fajlove()
```

```python
def ucitaj_linkove(kategorija):
    naziv_fajla = f"{kategorija.upper()}.txt"
    putanja = os.path.join(base_input_dir, naziv_fajla)

    with open(putanja, 'r', encoding='utf-8') as f:
        linkovi = [linija.strip() for linija in f if linija.strip() and linija.strip().startswith('http')]

    print(f"Učitano {len(linkovi)} linkova za kategoriju {kategorija}")
    return linkovi
```

```python
def parsiraj_clanak(url_kategorija_tuple):
    url, kategorija_rubrika = url_kategorija_tuple # Raspakiraj tuple

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Svaka nit sada radi svoj GET zahtjev
        page = requests.get(url, headers=headers, timeout=REQUEST_TIMEOUT)
        page.raise_for_status()
        html_content_raw = page.content
    except requests.exceptions.RequestException as e:
        #print(f"  Greška pri dohvaćanju stranice {url}: {e}") # Smanji verbose ispis u concurrent modu
        return None

    try:
        html_content_decoded = html_content_raw.decode('utf-8', errors='replace')
    except UnicodeDecodeError:
        #print(f"  Greška pri dekodiranju UTF-8 za {url}. Pokušavam s 'latin-1'.")
        try:
            html_content_decoded = html_content_raw.decode('latin-1', errors='replace') # Fallback encoding
        except Exception as e_decode:
            #print(f"  Potpuna greška pri dekodiranju za {url}: {e_decode}")
            return None


    # Pokušaj koristiti lxml ako je dostupan, inače html.parser
    parser_to_use = 'html.parser'
    try:
        import lxml
        parser_to_use = 'lxml'
    except ImportError:
        pass # lxml nije instaliran, koristi default html.parser

    soup = BeautifulSoup(html_content_decoded, parser_to_use)

    clanak_data = {
        "NOVINA": NOVINA_NAZIV,
        "DATUM": "N/A",
        "RUBRIKA": kategorija_rubrika,
        "NADNASLOV": "N/A",
        "NASLOV": "N/A",
        "PODNASLOV": "N/A",
        "STRANA": url,
        "AUTOR(I)": "N/A",
        "SADRZAJ": ""
    }
    author_from_pise_tag = None

    # LD+JSON obrada
    ld_json_script = soup.find('script', type='application/ld+json')
    if ld_json_script:
        try:
            ld_data = json.loads(ld_json_script.string)
            clanak_data["NASLOV"] = html.unescape(ld_data.get("headline", "N/A").strip())
            clanak_data["PODNASLOV"] = html.unescape(ld_data.get("description", "N/A").strip())

            date_published_str = ld_data.get("datePublished")
            if date_published_str:
                try:
                    dt_object = datetime.fromisoformat(date_published_str.replace("Z", "+00:00"))
                    clanak_data["DATUM"] = dt_object.strftime("%d.%m.%Y")
                except ValueError: # Fallback za druge formate datuma ako ISO ne uspije
                    try:
                        dt_object = datetime.strptime(date_published_str.split("T")[0], "%Y-%m-%d")
                        clanak_data["DATUM"] = dt_object.strftime("%d.%m.%Y")
                    except ValueError: pass # Ostavi N/A ako ni ovo ne uspije

            author_info = ld_data.get("author")
            if isinstance(author_info, dict):
                clanak_data["AUTOR(I)"] = html.unescape(author_info.get("name", "N/A").strip())
            elif isinstance(author_info, list) and author_info:
                authors_list = [html.unescape(a.get("name", "").strip()) for a in author_info if isinstance(a, dict) and a.get("name")]
                clanak_data["AUTOR(I)"] = ", ".join(filter(None, authors_list)).strip() or "N/A"
        except (json.JSONDecodeError, AttributeError, TypeError):
            pass # Tiho propusti ako JSON-LD nije validan ili nedostaje polje

    # Fallback na OG tagove ako LD+JSON nije dao podatke ili nije postojao
    if clanak_data["NASLOV"] == "N/A":
        og_title = soup.find('meta', property='og:title')
        if og_title and og_title.get('content'):
            clanak_data["NASLOV"] = html.unescape(og_title['content'].split('|')[0].strip())
    if clanak_data["PODNASLOV"] == "N/A":
        og_description = soup.find('meta', property='og:description') # ili name='description'
        if og_description and og_description.get('content'):
            clanak_data["PODNASLOV"] = html.unescape(og_description['content'].strip())
    if clanak_data["DATUM"] == "N/A":
        article_published_time = soup.find('meta', property='article:published_time')
        if article_published_time and article_published_time.get('content'):
            try:
                dt_object = datetime.fromisoformat(article_published_time['content'].replace("Z", "+00:00"))
                clanak_data["DATUM"] = dt_object.strftime("%d.%m.%Y")
            except ValueError: pass

    # Izvlačenje teksta članka
    editor_div = soup.find('div', class_=lambda x: x and 'editor' in x.split()) # Traži klasu koja SADRŽI 'editor'
    if editor_div:
        # Uklanjanje nepoželjnih elemenata
        for el_to_remove in editor_div.find_all(['script', 'style', 'noscript', 'form', 'iframe', 'figure', 'figcaption', 'aside']):
            el_to_remove.decompose()
        for ad_banner in editor_div.find_all(['div', 'section'], class_=re.compile(r'banner|ad-in-text|google|reklama|oglas|promo|social|widget', re.IGNORECASE)):
            ad_banner.decompose()
        for el in editor_div.find_all(string=re.compile(r'Reklama - nastavite čitati ispod oglasa|PR tekst', re.IGNORECASE)):
            if el.parent and el.parent.name != '[document]': # osiguraj da ima roditelja prije decompose
              el.parent.decompose()


        article_text_parts = []
        # Pronađi sve <p> tagove unutar editor_div. recursive=False može pomoći ako je struktura čista.
        # Za sada, uzimamo sve <p> pa čistimo.
        paragraphs = editor_div.find_all('p')

        # Ukloni duplikate paragrafa na temelju objekta (ako BeautifulSoup vraća iste objekte više puta)
        # Ovo je sofisticiranije od list(dict.fromkeys(clean_paragraphs)) koje radi samo na hashable elementima
        unique_paragraphs_obj = []
        seen_paragraph_ids = set()
        for p_obj in paragraphs:
            if id(p_obj) not in seen_paragraph_ids:
                unique_paragraphs_obj.append(p_obj)
                seen_paragraph_ids.add(id(p_obj))
        paragraphs = unique_paragraphs_obj

        for p_tag in paragraphs:
            # Provjera za ugniježđene <p> tagove - ako ovaj <p> sadrži drugi <p>, preskoči ga.
            # Ovo pomaže ako `recursive=False` nije korišteno ili nije dovoljno.
            if p_tag.find('p'):
                continue

            p_text_stripped = p_tag.get_text(separator=' ', strip=True)
            p_text_unescaped = html.unescape(p_text_stripped)
            p_text_lower = p_text_unescaped.lower()

            if not p_text_unescaped: # Preskoči potpuno prazne paragrafe
                continue

            # Logika za "Piše:"
            if p_text_lower.startswith("piše:"):
                potential_author = p_text_unescaped[len("piše:"):].strip()
                if potential_author and not any(char.isdigit() for char in potential_author): # Osnovna provjera da nije samo broj ili sl.
                    author_from_pise_tag = potential_author
                continue # Ne dodaj "Piše: ..." u sadržaj članka

            # Ignoriranje paragrafa na osnovu prefiksa
            skip_paragraph = False
            for prefix in PREFIXES_TO_IGNORE_IN_CONTENT:
                if p_text_lower.startswith(prefix):
                    skip_paragraph = True
                    break
            if skip_paragraph:
                continue

            # Dodatno čišćenje - npr. paragrafi koji su samo linkovi ili vrlo kratki promotivni tekstovi
            # Ovo je heuristika i treba je pažljivo testirati
            a_tags = p_tag.find_all('a')
            text_content_without_links = p_text_unescaped
            for a_tag in a_tags:
                text_content_without_links = text_content_without_links.replace(a_tag.get_text(strip=True), "")

            if len(text_content_without_links.strip()) < 15 and len(a_tags) > 0 : # Vrlo malo teksta osim linkova
                 if "klik" in p_text_lower or "opširnije" in p_text_lower or "pročitajte" in p_text_lower or "pogledajte" in p_text_lower :
                    continue

            article_text_parts.append(p_text_unescaped)

        # Uklanjanje duplikata stringova iz liste paragrafa
        unique_text_parts = []
        seen_text_parts = set()
        for part in article_text_parts:
            if part not in seen_text_parts:
                unique_text_parts.append(part)
                seen_text_parts.add(part)

        full_text = " ".join(unique_text_parts)
        clanak_data["SADRZAJ"] = re.sub(r'\s+', ' ', full_text).strip()

    # Primijeni autora iz "Piše:" ako je bolji
    if author_from_pise_tag:
        current_author_lower = clanak_data["AUTOR(I)"].lower()
        is_current_author_generic = any(ga in current_author_lower for ga in GENERIC_AUTHOR_NAMES)

        if is_current_author_generic or clanak_data["AUTOR(I)"] == "N/A":
            clanak_data["AUTOR(I)"] = html.unescape(author_from_pise_tag)
        # Može se dodati i logika spajanja autora ako su oba validna i različita

    # Fallback za naslov ako i dalje N/A
    if clanak_data["NASLOV"] == "N/A" or not clanak_data["NASLOV"].strip():
        title_tag_html = soup.find('title')
        if title_tag_html and title_tag_html.string:
             clanak_data["NASLOV"] = html.unescape(title_tag_html.string.split('|')[0].strip())

    # Finalno unescape za ključna polja za svaki slučaj (iako bi trebalo biti pokriveno)
    for key in ["NASLOV", "PODNASLOV", "AUTOR(I)"]:
        if isinstance(clanak_data[key], str):
            clanak_data[key] = html.unescape(clanak_data[key].strip())
        elif clanak_data[key] is None: # Osiguraj da nema None vrijednosti, zamijeni s N/A
             clanak_data[key] = "N/A"


    if not clanak_data["SADRZAJ"].strip() and (clanak_data["NASLOV"] == "N/A" or not clanak_data["NASLOV"].strip()):
        #print(f"    Članak {url} nema validan naslov ni sadržaj. Preskačem.")
        return None
    if not clanak_data["NASLOV"].strip(): # Ako je naslov prazan string, postavi na N/A
        clanak_data["NASLOV"] = "N/A"

    return clanak_data
```

```python
def formatiraj_za_txt(clanak_data):
    txt_output_lines = [
        "<***>",
        f"NOVINA: {clanak_data.get('NOVINA', 'N/A')}",
        f"DATUM: {clanak_data.get('DATUM', 'N/A')}",
        f"RUBRIKA: {clanak_data.get('RUBRIKA', 'N/A')}",
        f"NADNASLOV: {clanak_data.get('NADNASLOV', 'N/A')}",
        f"NASLOV: {clanak_data.get('NASLOV', 'N/A')}",
        f"PODNASLOV: {clanak_data.get('PODNASLOV', 'N/A')}",
        f"STRANA: {clanak_data.get('STRANA', 'N/A')}",
        f"AUTOR(I): {clanak_data.get('AUTOR(I)', 'N/A')}",
        "", # Prazan red nakon metapodataka
        clanak_data.get('SADRZAJ', '').strip(), # Tekst članka, osiguraj da nema vodećih/pratećih razmaka
        ""  # Prazan red nakon teksta članka (prije sljedećeg <***> ako se dodaje van ove funkcije)
    ]
    return "\n".join(txt_output_lines)
```

```python
def obradi_i_sacuvaj_kategoriju(kategorija_obrada, target_date_obj=None):
    print(f"\n➤ Počinjem obradu kategorije: {kategorija_obrada}")

    linkovi = ucitaj_linkove(kategorija_obrada)
    if not linkovi:
        print(f"Nema linkova za obradu u kategoriji {kategorija_obrada}.")
        return

    output_folder_path = os.path.join(base_output_dir, kategorija_obrada)
    txt_output_path = os.path.join(output_folder_path, f"{kategorija_obrada}.txt")
    json_output_path = os.path.join(output_folder_path, f"{kategorija_obrada}.json")

    svi_clanci_za_kategoriju_json = []

    # Pripremi argumente za svaku nit - tuple (url, kategorija_obrada)
    tasks_args = [(url, kategorija_obrada) for url in linkovi]

    processed_count = 0
    start_time_cat = time.time()

    with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS_THREADS) as executor:
        # Koristi map za izvršavanje `parsiraj_clanak` za svaki set argumenata
        # `parsiraj_clanak` sada prima tuple `(url, kategorija_obrada)`
        future_to_url = {executor.submit(parsiraj_clanak, args): args[0] for args in tasks_args}

        for future in concurrent.futures.as_completed(future_to_url):
            url_original = future_to_url[future]
            try:
                clanak_data = future.result() # Dobij rezultat (ili izuzetak ako ga je bilo)
                processed_count += 1

                if processed_count % 50 == 0: # Ispis napretka svakih 50 obrađenih linkova
                    elapsed_time = time.time() - start_time_cat
                    print(f"   Kategorija {kategorija_obrada}: Obrađeno {processed_count}/{len(linkovi)} linkova. Vrijeme: {elapsed_time:.2f}s")

                if clanak_data:
                    if target_date_obj and clanak_data["DATUM"] != "N/A":
                        try:
                            datum_clanka = datetime.strptime(clanak_data["DATUM"], "%d.%m.%Y")
                            if datum_clanka >= target_date_obj:
                                #print(f"    Članak '{clanak_data.get('NASLOV', 'Nepoznat naslov')}' ({clanak_data['DATUM']}) je na ili nakon target_date. Preskačem.")
                                continue
                        except ValueError:
                            #print(f"    Greška pri parsiranju datuma '{clanak_data['DATUM']}' za provjeru target_date.")
                            pass # Nastavi bez filtriranja ako datum nije parsiran

                    svi_clanci_za_kategoriju_json.append(clanak_data)
            except Exception as exc:
                #print(f"  Greška pri obradi URL-a {url_original} u niti: {exc}")
                pass # Tiho nastavi ako nit pukne za jedan URL

    end_time_cat = time.time()
    print(f"✔ Završeno parsiranje za kategoriju {kategorija_obrada}. Ukupno prikupljeno članaka: {len(svi_clanci_za_kategoriju_json)}. Vrijeme obrade: {end_time_cat - start_time_cat:.2f}s")

    if svi_clanci_za_kategoriju_json:
        # Sortiraj članke po datumu (ako postoji i ako je validan) prije spremanja, opcionalno
        # Ovo može biti korisno za konzistentnost, ali dodaje malo vremena obrade
        def get_date_for_sort(article):
            try:
                return datetime.strptime(article.get("DATUM", "01.01.1900"), "%d.%m.%Y")
            except ValueError:
                return datetime(1900,1,1) # Za članke bez datuma ili s nevalidnim datumom

        svi_clanci_za_kategoriju_json.sort(key=get_date_for_sort, reverse=True) # Najnoviji prvo, na primjer


        try:
            with open(json_output_path, 'w', encoding='utf-8') as json_file:
                json.dump(svi_clanci_za_kategoriju_json, json_file, ensure_ascii=False, indent=4)
            print(f"  Svi članci za kategoriju {kategorija_obrada} spremljeni u {json_output_path}")
        except IOError as e:
            print(f"  Greška pri spremanju JSON datoteke {json_output_path}: {e}")
            return

        try:
            with open(txt_output_path, 'w', encoding='utf-8') as txt_file:
                for idx, clanak_entry in enumerate(svi_clanci_za_kategoriju_json):
                    txt_representation = formatiraj_za_txt(clanak_entry)
                    txt_file.write(txt_representation)
                    if idx < len(svi_clanci_za_kategoriju_json) - 1:
                         txt_file.write("\n")
            print(f"  Svi članci za kategoriju {kategorija_obrada} formatirani i spremljeni u {txt_output_path}")
        except IOError as e:
            print(f"  Greška pri spremanju TXT datoteke {txt_output_path}: {e}")
    else:
        print(f"Nema članaka za spremanje za kategoriju {kategorija_obrada}.")
```

```python
import pdfplumber
import re
from langdetect import detect
import unicodedata
import json
```

```python
# Učitaj PDF
pdf_path = "/content/drive/My Drive/CasopisProzor/CasopisProzor.pdf"
articles_raw = []
current_article = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if not text:
            continue
        lines = text.split("\n")
        for line in lines:
            if re.fullmatch(r'\d{1,3}', line.strip()):
                continue
            if line.startswith("UDK"):
                if current_article:
                    articles_raw.append(current_article.strip())
                current_article = line
            else:
                current_article += "\n" + line
    if current_article:
        articles_raw.append(current_article.strip())
```

```python
# Funkcija za prepoznavanje ćirilice
def is_cyrillic(text):
    return any('CYRILLIC' in unicodedata.name(c) for c in text if c.isalpha())
```

```python
def extract_metadata_and_text_v3(text, start_page=None, end_page=None):
    metadata = {
        "NOVINA": "Prozor u svijet obrazovanja, nauke i mladih",
        "DATUM": "N/A",
        "RUBRIKA": "N/A",
        "NADNASLOV": "N/A",
        "NASLOV": "N/A",
        "PODNASLOV": "N/A",
        "STRANA": f"{start_page + 1}-{end_page + 1}" if start_page is not None and end_page is not None else "N/A",
        "AUTOR(I)": "N/A"
    }

    lines = [line.strip() for line in text.split('\n')]
    if len(lines) < 10 or not lines[0].startswith("UDK"):
        return None, None

    # Ekstrakcija autora iz blokova razdvojenih praznim linijama
    i = 4
    author_lines = []
    author_buffer = []
    while i < len(lines) and not lines[i].startswith(("Lektor", "Korektura")):
        line = lines[i].strip()
        if line == "":
            if author_buffer:
                for l in author_buffer:
                    if l:
                        author_lines.append(l)
                        break
                author_buffer = []
        else:
            author_buffer.append(line)
        i += 1
    if author_buffer:
        for l in author_buffer:
            if l:
                author_lines.append(l)
                break

    metadata["AUTOR(I)"] = ', '.join(author_lines)

    while i < len(lines) and lines[i].startswith(("Lektor", "Korektura")):
        i += 1
    while i < len(lines):
        if lines[i]:
            metadata["NASLOV"] = lines[i].strip()
            i += 1
            break
        i += 1

    uvod_pattern = re.compile(
        r"^(1\.*\s*)?(uvod|teorijski okvir rada|uvodna razmatranja|analiza koncepta.*|povijesni pogled.*|igra)",
        re.IGNORECASE
    )
    while i < len(lines):
        if uvod_pattern.match(lines[i]):
            i += 1
            break
        i += 1

    body = []
    literatura_pattern = re.compile(
        r"^(\d{0,2}\.?\s*)?(literatura|literatura preuzeta s interneta)\:?",
        re.IGNORECASE
    )
    while i < len(lines):
        line = lines[i].strip()
        if literatura_pattern.match(line):
            break
        if is_cyrillic(line):
            i += 1
            continue
        try:
            if detect(line) != 'hr':
                i += 1
                continue
        except:
            i += 1
            continue
        body.append(line)
        i += 1

    full_body = ' '.join(body).strip()
    return metadata, full_body
```

```python
# Priprema članaka
articles_cleaned = []
for a in articles_raw:
    meta, body = extract_metadata_and_text_v3(a)
    if not body:
        continue
    articles_cleaned.append({"metadata": meta, "body": body})
```

```python
# Spašavanje u JSON i TXT
with open("casopis_prozor_v3.json", "w", encoding="utf-8") as jf:
    json.dump(articles_cleaned, jf, ensure_ascii=False, indent=2)

with open("casopis_prozor_formatiran_v3.txt", "w", encoding="utf-8") as tf:
    for article in articles_cleaned:
        tf.write("<***>\n")
        for key, value in article["metadata"].items():
            tf.write(f"{key}: {value}\n")
        tf.write("\n")
        tf.write(article["body"])
        tf.write("\n\n")
```
